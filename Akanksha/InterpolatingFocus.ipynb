```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
from scipy import interpolate
```

```python
dataframe = pd.read_excel("Focus2006.xlsx")
```

```python
#Interpolation function(Modified Julian and Focus are given)
focus_fcn = interpolate.interp1d(dataframe['Julian '], dataframe['Focus'],)
focus_oct_fcn = interpolate.interp1d(dataframe['Julian '][(dataframe['Julian ']>54009.00000000) & (dataframe['Julian ']<54039.9999999)], dataframe['Focus'][(dataframe['Julian ']>54009.00000000) & (dataframe['Julian ']<54039.9999999)])
```

```python
#modied julian dates with interval of 10 mins, the exposure period is about 2088 seconds. 
#The first value of array is the start time from https://mast.stsci.edu/portal/Mashup/Clients/Mast/Portal.html
oct_vals = np.array([54010.79328704, 54010.80023148, 54010.80717593, 54010.81412037])
focus_oct_vals = focus_fcn(oct_vals)
#focus_oct_fcn_vals = focus_oct_fcn(oct_vals)
```

```python
%matplotlib notebook
plt.plot(dataframe['Julian '], dataframe['Focus'], '.')
plt.plot(oct_vals, focus_oct_vals, '.')
```

```python
#Trying to check the mean focus for september and oct
mean_focus_sept = np.mean(dataframe['Focus'][(dataframe['Julian ']>53979.00000000) & (dataframe['Julian ']<	54008.999999)])
mean_focus_oct = np.mean(dataframe['Focus'][(dataframe['Julian ']>54009.00000000) & (dataframe['Julian ']<54039.9999999)])
```
